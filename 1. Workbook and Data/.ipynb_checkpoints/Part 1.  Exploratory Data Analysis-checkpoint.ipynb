```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
```

```python
df = pd.read_json('logins.json')
df.head()
```

Start by importing the necessary tools an data set.

```python
df.info()
```

```python
df['login_time'] = pd.to_datetime(df['login_time'])
df.head()
```

Make sure the format is converted to date time.

```python
df.set_index('login_time', inplace=True)
df['# of logins'] = 1
df.head()
```

Set the date time as the index

```python
df_15 = df.resample('15min').sum()
df_15.head()
```

Make the time iterate over 15 minute intervals

```python
def show_na(df):
    na_df = pd.concat([df.isnull().sum(), 100 * df.isnull().mean()], axis=1)
    na_df.columns=['count', '%']
    na_df.sort_values(by='count', ascending = False)
    return na_df
# Use the function
show_na(df_15)
```

Make sure there isn't any null entries.

```python
plt.figure(figsize = (18,6))
plt.plot(df_15, color='green')
plt.title('Logins per 15-min Interval')
plt.ylabel('# of Logins')
plt.xlabel('Date and Time')
```

This is a pretty congested plot. Using rolling averages will probably look cleaner and show the averages.

```python
df_15roll = df_15.rolling(window=96).mean()
plt.figure(figsize = (18,6))
plt.plot(df_15roll, color='green')
plt.title('Logins per Rolling Window of 1 Day ')
plt.ylabel('# of Logins')
plt.xlabel('Date and Time')
plt.show()
```

There is a  large peak in April, perhaps Easter fell on that day in 1970.

In order to see the rolling average for each day:

15 minutes in an  hour (15 * 4 = 60),

24  hours in a day (4 * 24 =  96).

Therefore, a window of 96 produces the average logins per day, making the data much easier to visualize. When working with timeseries, though, it is important to check for stationarity in the data. This helps in evaluating the best model for forcasting the future trends. 

```python
from statsmodels.tsa.stattools import adfuller
def test_stationarity(timeseries):

    #Determing rolling statistics
    rolmean = timeseries.rolling(96).mean()
    rolstd = timeseries.rolling(96).mean()

    #Plot rolling statistics:
    fig = plt.figure(figsize=(12, 8))
    orig = plt.plot(timeseries, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show()
    
    #Perform Dickey-Fuller test:
    print('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print(dfoutput) 
```

```python
test_stationarity(df_15)
```

```python
df_15['first_diff'] = df_15 - df_15.shift(2)
test_stationarity(df_15['first_diff'].dropna(inplace=False))
```

```python
date = dt.datetime(1970, 1, 1)
date.weekday()
```

```python
df_1week = df_15.loc['1970-01-05':'1970-01-19'].rolling(window=96).mean()
plt.figure(figsize = (18,6))
plt.plot(df_1week, color='green')
plt.title('Logins per Rolling Window of 1 Day ')
plt.ylabel('# of Logins')
plt.xlabel('Date and Time')
plt.show()
```

We can see above that the timeseries is stationary, and there appears to be a clear trend of more logins at the end of the week, mainly on Fridays.
